```python

class VehicleSize(Enum):

    motorcycle = 0
    compact = 1
    large = 2
class parkinglot:
    
    def __init__(self,num_levels):
        self.num_levels = num_levels
        self.levels = []
    def park_veichle(self,veichle):
        pass
class level:
    def __init__(self,floor,total_spots):
        self.floor = floor
        self.total_spots = total_spots
        self.available = 0
        self.parking_spots = []
    def park_veichle(self,veichle):
        pass
    def spot_free(self):
        pass
    def find_available_spot(self,veichle):
        pass
class parkingspot:
    def __init__(self,floor,row,spot_id,spot_size,veichle_size):
        self.floor = floor
        self.row= row
        self.spot_id = spot_id
        self.spot_size = spot_size
        self.veichle_size = veichle_size
        self.veichle = None
    def is_available(self):
        return self.veichle==None
    def can_fit(self,veichle):
        if self.veichle==None and veichle.can_fit(self):
            return True
        return False
    def park_veichle(self, veichle):
        self.veichle = veichle
    def remove_veichle(self):
        self.veichle =None
class veichle:
    def __init__(self,model,year,veichle_size,plate,spot_size):
        self.model = model
        self.year = year
        self.plate = plate
        self.veichle_size = veichle_size
        self.spot_size = spot_size
        self.spots_taken =[]
    def take_spot(self, parkingspot):
        self.spots_taken.append(parkingspot)
    def clear_spot(self):
        for spot in self.spots_taken:
            spot.remove_vehicle(self)
        self.spots_taken = []
class motorcycle(veichle):
    def __init__(self,model,year,plate):
        super().__init__(model,year,plate,veichle_size=VehicleSize.motorcycle,spot_size=1)
    def can_fit(self,parkingspot):
        return True
class car(veichle):
    def __init__(self,model,year,plate):
        super().__init__(model,year,plate,veichle_size=VehicleSize.compact,spot_size=1)
    def can_fit(self,parkingspot):
        return True if (parkingspot.veichle_size==compact or parkingspot.veichle_size==large) else False
class truck(veichle):
    def __init__(self,model,year,plate):
        super().__init__(model,year,plate,veichle_size=VehicleSize.large,spot_size=1)
    def can_fit(self,parkingspot):
        return True if (parkingspot.veichle_size==large) else False
class bus(veichle):
    def __init__(self,model,year,plate):
        super().__init__(model,year,plate,veichle_size=VehicleSize.large,spot_size=5)
    def can_fit(self,parkingspot):
        return True if (parkingspot.veichle_size==large) else False
```
